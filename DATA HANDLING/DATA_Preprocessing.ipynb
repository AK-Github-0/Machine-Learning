```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("train.csv")
df1.head()
```

```python
df2 = pd.read_csv("census_starter.csv")
df2
```

```python
df = pd.merge(df1, df2, left_on="cfips", right_on="cfips")
df
```

# JUST extracting data row by row to check which line should be added from the other file

```python
DL = []
SL = []
year = []
for i in df.iterrows():
    name = str(i[1]["first_day_of_month"]).split("-")[0]
    SL.append(i[1]["row_id"])
    SL.append(name)
    for j in df2.columns:
        k = j.split("_")[-1]
        year.append(k)
        if k == name:
            SL.append(i[1][j])
    DL.append(SL)
    SL = []
DL
```

```python
df3 = pd.DataFrame(data =DL, columns=["row_id", "Year", "pct_bb", "pct_college", "pct_foregin", "pct_it", "median_hh_inc"])
df3
```

```python
for i in df3.columns:
    if i == "row_id":
        continue
    df3[i].fillna(np.mean(df3[i]), inplace=True)
```

```python
df = pd.merge(df1, df3, left_on="row_id", right_on="row_id")
df
```
